### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
df = pd.read_csv(url, sep='\t')
df
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
```

### Step 4. How many products cost more than $10.00?

```python
#chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
#chipo['item_price'] = pd.to_numeric(chipo['item_price'].astype(str).str[1:])
b = lambda x: float(x[1:-1])
chipo['item_price'] = chipo['item_price'].apply(b)
print(chipo['item_price'])
products_over_10 = chipo[chipo['item_price'] > 10.00]

num_products_over_10 = len(products_over_10)

print(num_products_over_10)
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
item_prices = chipo[['item_name', 'item_price']]

unique_item_prices = item_prices.drop_duplicates()

print(unique_item_prices)
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by=['item_name'], inplace=True)
chipo
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# Số lượng mặt hàng đắt nhất được đặt hàng là bao nhiêu?
most_expensive_item_price = chipo['item_price'].max()
most_expensive_item = chipo[chipo['item_price'] == most_expensive_item_price]

quantity_most_expensive = most_expensive_item['quantity'].values[0]
name_most_expensive = most_expensive_item['item_name'].values[0]

print("Tên món hàng đắt nhất là: ",name_most_expensive)
print("\nGiá món hàng đắt nhất là: ",most_expensive_item_price)
print("\nSố lượng đặt hàng của món đắt nhất là: ",quantity_most_expensive)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
num_veggie_salad_bowl_orders = len(veggie_salad_bowl_orders)

print("Veggie Salad Bowl đươc đặt: ",num_veggie_salad_bowl_orders)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
num_canned_soda_orders = len(canned_soda_orders)

print("Số lần đặt nhiều hơn 1 món Canned Soda là: ",num_canned_soda_orders)
```
