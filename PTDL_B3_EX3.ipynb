# US - Baby Names

### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv).

```python
data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv')
data
```

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = data
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
baby_names
```

### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names.groupby('Gender')['Count'].sum()
print(gender_counts)

if gender_counts['M'] > gender_counts['F']:
    print('Có nhiều tên nam hơn tên nữ.')
elif gender_counts['M'] < gender_counts['F']:
    print('Có nhiều tên nữ hơn tên nam.')
else:
    print('Số lượng tên nam và nữ bằng nhau.')
```

### Step 7. Group the dataset by name and assign to names

```python
#names = baby_names.groupby('Name')
names = baby_names.groupby('Name').agg({'Count': 'sum', 'Year': 'first', 'Gender': 'first', 'State': 'first'})
names
```

### Step 8. How many different names exist in the dataset?

```python
num_different_names = names.index.nunique()
print("Số tên khác nhau là: ", num_different_names)
```

### Step 9. What is the name with most occurrences?

```python

most_name = names['Count'].idxmax()
print("Tên xuất hiện nhiều nhất là: ", most_name)
```

### Step 10. How many different names have the least occurrences?

```python

min_count = names['Count'].min()

least_frequent_names = names[names['Count'] == min_count]
print(least_frequent_names)

num_least_frequent_names = len(least_frequent_names)
print("Số lượng tên ít xuất hiện là: ", num_least_frequent_names)
```

### Step 11. What is the median name occurrence?

```python
median_name = names['Count'].median()
print("Số lần xuất hiện tên trung bình là: ", median_name)
```

### Step 12. What is the standard deviation of names?

```python
std_name = names['Count'].std()
print("Độ lệch chuẩn của tên là: ", std_name)
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
summary = names['Count'].describe()
print(summary)
```
